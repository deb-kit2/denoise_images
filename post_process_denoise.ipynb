```
from skimage.io import imread, imshow, imsave
from skimage import exposure

res1 = imread("/content/new1.jpg")
res2 = imread("/content/new2.jpg")
```

##gamma

```
imshow(exposure.adjust_gamma(res1, 4, 1))
```

```
imsave("new1gamma.jpg", exposure.adjust_gamma(res1, 4, 1))
imsave("new2gamma.jpg", exposure.adjust_gamma(res2, 4, 1))
```

```
imsave("res2gamma35.jpg", exposure.adjust_gamma(res1, 3.5, 1))
imsave("res2gamma40.jpg", exposure.adjust_gamma(res1, 3.5, 1))
```

##log : not using

```
imshow(exposure.adjust_log(res1, 1, True))
```

```
#imsave("res1log.jpg", exposure.adjust_log(res1, 4, False))
#not good
```

```
imsave("res1loginv.jpg", exposure.adjust_log(res1, 2, True))
```

##sigmoid

```
imshow(exposure.adjust_sigmoid(res1, cutoff = 0.85, gain = 18, inv = False))
```

```
imsave("new1sig.jpg", exposure.adjust_sigmoid(res1, cutoff = 0.85, gain = 18, inv = False))
```

```
imsave("new2sig.jpg", exposure.adjust_sigmoid(res2, cutoff = 0.85, gain = 18, inv = False))
```

##rescale_intensity

```
imshow(exposure.rescale_intensity(res1, in_range = (150, 255), out_range = "dtype"))
```

```
imsave("res1rescaleint.jpg", exposure.rescale_intensity(res1, in_range = (150, 255), out_range = "dtype"))
```

```
imsave("res2rescaleint.jpg", exposure.rescale_intensity(res2, in_range = (150, 255), out_range = "dtype"))
```

##equalize_adapthist

```
imshow(exposure.equalize_adapthist(res1, None, clip_limit = 0.01))
```

```
imsave("res1adapthist.jpg", exposure.equalize_adapthist(res1, None, clip_limit = 0.01))
```

```
imsave("res2adapthist.jpg", exposure.equalize_adapthist(res2, None, clip_limit = 0.01))
```
